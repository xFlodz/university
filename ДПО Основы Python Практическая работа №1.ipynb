#### 1. Запросите у пользователя три разных числа и выведите их произведение.


````python
list_of_nums = []
for i in range(3):
  num = int(input(f'Введите число №{i+1} '))
  list_of_nums.append(num)

answer = 1
for num in list_of_nums:
  answer *= num

print(answer)
````

#### 2. Попросите пользователя ввести свое ФИО, а затем выведите фамилию и инициалы.


````python
def fullname_formate():

  fullname = input('Введите ФИО: #Иванов Иван Иванович ').strip().split(' ')

  if len(fullname) < 2 or len(fullname) > 3:
    return 'Вы неверно ввели имя'

  list_of_words = [i.capitalize() for i in fullname]

  if len(list_of_words) == 3:
    return f'{list_of_words[0]} {list_of_words[1][0]}. {list_of_words[2][0]}.'

  else:
    return f'{list_of_words[0]} {list_of_words[1][0]}.'

print(fullname_formate())
````
